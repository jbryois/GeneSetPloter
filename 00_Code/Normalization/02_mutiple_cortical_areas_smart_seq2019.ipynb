```python
import pandas as pd
import numpy as np
```

```python
file_name='../../00_Raw/mutiple_cortical_areas_smart_seq2019/matrix.csv'
meta_data='../../00_Raw/mutiple_cortical_areas_smart_seq2019/metadata.csv'
```

```python
meta=pd.read_csv(meta_data)
```

```python
meta = meta[['sample_name','cluster_label']]
```

```python
meta.head()
```

```python
meta_dict = dict(zip(meta.sample_name,meta.cluster_label))
```

```python
expression_sum = {}
with open(file_name) as f:
    
  #get gene labels from header
  genes = f.readline().split(',')[1:]

  #for each line get cell type identity of the single cell and sum the total counts
  for count,line in enumerate(f):
    line_elements = line.split(',')
    cell_id = line_elements[0]
    
    #Get cell type identity corresponding to cell id
    cell_type_id = meta_dict[cell_id]
    
    #Get expression of all genes in a numpy array
    expression = np.array([int(i) for i in line_elements[1:]])
    
    #Create new cell type id - expression dictionary if key is not present
    if cell_type_id not in expression_sum:
        expression_sum[cell_type_id] = expression
    #Else add the counts to the cell id
    else:
        expression_sum[cell_type_id] = np.add(expression_sum[cell_type_id], expression)
    if count % 1000 ==0:
        print(count)
```

```python
expression_sum_df = pd.DataFrame(expression_sum,index=genes)
```

```python
expression_sum_df.index.name = 'Gene'
```

```python
expression_sum_df = expression_sum_df.loc[:, expression_sum_df.columns.notnull()]
```
